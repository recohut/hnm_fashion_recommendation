# Changing all to lowercase

Changing all strings to lowercase to check for duplications

```python
# lowercase everything then check for duplication
for col in articles.columns:
    if articles[col].dtype == 'object':
        articles[col] = articles[col].str.lower()
```

```python
articles.head(2)
```

# Removing duplicates

```python
articles_dup = articles[articles.iloc[:,1:].duplicated()]
articles_dup.head()
```

```python
articles[articles['product_code']==145872]
```

```python
article_dup_id = articles_dup['article_id'].to_list()
```

```python
articles2 = articles[~articles['article_id'].isin(article_dup_id)].reset_index()
```
